# Reading Resumes and Sentence Segmentation

- in this cell the resume list was used to access the resumes using a for loop
- all the resumes were appended into a list
- using nltk's sent_tokenize library sentences were segmented
- all the sentences of the resume were saved into a list "sentences"


```python
import nltk
nltk.download('punkt')
nl=[]
resumes=""
for i in range (0,len(lis)-1):
    string="resume_txt/resume_("+str(lis[i])+").txt"
    #print(string)
    db= open(string,'r',errors='ignore',encoding="utf-8")
    x=db.read()
    nl.append(x)
for resume in nl:
    resumes+=resume
    
#print(nl) 
print(resumes)
##sentence segmentation
from nltk.tokenize import sent_tokenize
sentences=sent_tokenize(resumes)
#print(sentences)
    
```

# Word Tokenization using Regex

- in this cell using the RegexpTokenizer we tokenize the words
- given regex was used \w+(?:[-']\w+)?
- by using the set we remove the repetetion in the list


```python
from nltk.tokenize import RegexpTokenizer 
tokenize = RegexpTokenizer(r"\w+(?:[-']\w+)?")
unigram = tokenize.tokenize(resumes)
#print (unigram)
voc=list(set(unigram))
print(voc)
```

# Selecting meaningful Bigrams

- in this cell we import MWETokenizer
- using this library we create bigram tokens

```python
from nltk.tokenize import MWETokenizer
bigrams = MWETokenizer(voc)
b_t = bigrams.tokenize(unigram)
#print (b_t)
```

- find meaningful bigrams
- removing numbered words form the tokenlist
- removing words with length less than 3 to find meaningful words
- appending the first 200 bigrams into a list

```python

#first 200 meaningful bigrams
bigram_list=[]
bi_2 = []
from nltk import bigrams
bgm=list(bigrams(b_t))
#print(bgm)
for i in range (0,len(bgm)-1):
    if not bgm[i][0].isdigit()  and not bgm[i][1].isdigit():
        #print(bgm[i][0])
        if len(bgm[i][0])>3 and len(bgm[i][1])>3:
            bigram_list.append(bgm[i])
for i in range(0,200):
    bi_2.append(bigram_list[i])
print(bi_2)    
```

# Lowercasing the letters in the sentence.

- check if the first letter in statement is uppercase
- if the second letter is uppercase then do nothing because it can be an abbrivation
- convert the first letter to lower case

```python
#lowercasing letters
lower=[]
for each in sentences:
    if each[0].isupper():
        if each[1].isupper():
            continue
        else:
            #print(each[0])
            lower.append(each.replace(each[0],each[0].casefold()))
print(lower)        
        
```

# refining wordlength , removing words with less than 3 characters

- a regex has been made to check wether the token is alphabet only and length is set from 3 to 45
- this regex will remove all unwanted numbers and words will have length from 3 to 45 as 45 is the largest recorded word in dictionary 
- after refining the tokens append them into a new list

```python
#refining string removing <3 and unwanted numbers
ns=[]
reg=''
reg=re.compile(r'^[a-zA-Z]{3,45}$',re.DOTALL)
for m in voc:
    #print(m)
    if re.search(reg,m):
        ns.append(m)
#
ref_voc=ns
print(ref_voc)
       
    
```

# Removing Stopwords

- stopwords english library is used
- with for loop check if the token exists in stopwords if not then append into the list

```python
from nltk.corpus import stopwords
swl = stopwords.words('english')
sws = set(swl)
st_voc = [w for w in ref_voc if w not in sws]
print (st_voc)
```

# removing rare tokens from the token list

- first get a dictionary with word count of every token in the resume
- update the dictionary 

```python
wf={}
for word in st_voc:
    c=0
    for res in nl:
        if word in res:
            c+=1
    if c!=0:
        wf.update({word:c})
print(wf)    
                
```
